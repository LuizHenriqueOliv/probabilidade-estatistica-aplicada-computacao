```python
import pandas as pd

df = pd.read_csv('data/fifa_players.csv')
```

Tabela de frequência para jogadores de cada nacionalidade

```python
freq = df["nationality"].value_counts()
limite = 300
maiores = freq[freq > limite]
outros = freq[freq <= limite].sum()
maiores['outros'] = outros
maiores
```

Gráfico de Barras para jogadores de cada nacionalidade

```python
maiores.columns = ['nacionalidade', 'jogadores']
maiores.plot(kind="bar")
```

```python
# criando uma tabela para um dado numérico
# jogadores com overall maior que 80
overall = df[df["overall_rating"] > 80]
overall = overall.overall_rating.value_counts()
overall.plot.bar()
```
